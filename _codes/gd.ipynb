```python
import numpy as np
from mlreg import SGDLinearRegression
```

```python
import matplotlib.pyplot as plt
```

```python
X = np.random.randn(1000, 2)
y = 3*X[:, 0]+2*X[:, 1]+1+np.random.randn(1000)
```

```python
mlr3 = SGDLinearRegression(
    learning_rate=0.01, num_iterations=1000, batch_size=5)
theta, _ = mlr3.fit(X, y)
mse_losses = mlr3.mse_losses()
plt.plot(mse_losses)
```
