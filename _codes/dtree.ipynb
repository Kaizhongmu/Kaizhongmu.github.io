```python
import matplotlib.pyplot as plt

# Data (x0, x1, and Class labels)
data = [
    [2, 3, 0], [3, 4, 0], [4, 6, 0], [6, 8, 1], [7, 10, 1],
    [8, 12, 1], [5, 7, 1], [2, 5, 0], [9, 15, 1], [1, 2, 0],
    [11, 3, 0], [4, 13, 1], [8, 14, 1], [1, 5, 0], [6, 2, 1],
    [9, 3, 1], [15, 13, 0], [7, 5, 0], [5, 9, 0], [8, 3, 1]
]

# Extracting x0, x1, and class labels
x0 = [row[0] for row in data]
x1 = [row[1] for row in data]
classes = [row[2] for row in data]

# Assigning colors based on class labels
colors = ['blue' if c == 0 else 'red' for c in classes]

# Plotting
plt.figure(figsize=(7, 5))
plt.grid(True)

# Plot points with filled circles
plt.scatter(x0, x1, color=colors, s=100, edgecolor='black')

# Label points with class values
for i in range(len(x0)):
    plt.text(x0[i] + 0.2, x1[i] + 0.2, str(classes[i]), fontsize=9)

# Set limits for the axes
plt.xlim(0, 16)
plt.ylim(0, 16)

# Label axes and show plot
plt.xlabel('$x_0$')
plt.ylabel('$x_1$')
plt.title('Scatter Plot of $x_0$ vs $x_1$ with Class Labels (Filled Circles)')
plt.show()
```
